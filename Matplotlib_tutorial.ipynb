## Scatter plot

Description: This type of plots are used to see the relationship between variables, and how changes can affect the other. Any of the variables that already have an assigned value, that is their default value, and if you do not add anything, the plot still works.


Syntaxis:

`matplotlib.pyplot.scatter(x_axis_data, y_axis_data, s=None, c=None, marker=None, cmap=None, vmin=None, vmax=None, alpha=None, linewidths=None, edgecolors=None) `

- **x_axis_data:** An array containing x-axis data
- **y_axis_data:** An array containing y-axis data
- **s:** marker size (the size of the points in the canvas)
- **c:** color of sequence of colors for markers
- **marker:** marker style
- **cmap:** cmap name
- **linewidths:** width of marker border
- **edgecolor:** marker border color
- **alpha:** blending value, between 0 (transparent) and 1 (opaque)

## How to use subplots

Syntaxis:

`matplotlib.pyplot.subplots(nrows=1, ncols=1, sharex=False, sharey=False, squeeze=True, subplot_kw=None, gridspec_kw=None, **fig_kw)`

As we see most all of the parameters have defautl instructions, so if we just call the function, it should work on its own.

Parameters:
- **nrows, ncols:** These parameter are the number of rows/columns of the subplot grid.
- **sharex, sharey:** These parameter controls sharing of properties among x (sharex) or y (sharey) axes.
- **squeeze:** This parameter is an optional parameter and it contains boolean value with default as True.
- **num:** This parameter is the pyplot.figure keyword that sets the figure number or label.
- **subplot_kwd:** This parameter is the dict with keywords passed to the add_subplot call used to create each subplot.
- **gridspec_kw:** This parameter is the dict with keywords passed to the GridSpec constructor used to create the grid the subplots are placed on.

## Distribution plots


Distribution plots are useful for assesing the probability distribution of our data. Why is that important?

```python
bc_data = pd.read_csv('bc_data.csv')
bc_data.head(5)
```

```python
bc_data.columns
```

We start with histograms. This plot is the graphical representation of the distribution of a variable. The x axis representes the ranges of the bins, while the Y axis reflects the frecuency.

Syntaxis:

`plt.hist(x, bins, density, range, histtype, align, weights, bottom, rwidth, color, label, log)`

<font color='red'>OP</font>: Optional parameter

- **x:** array or sequennce of array
- **bins:** <font color='red'>OP</font> that contains an integer or sequence or strings
- **density:** <font color='red'>OP</font> containgin boolean values
- **range:** <font color='red'>OP</font> representes the uper and lower range of bins
- **histtype:** <font color='red'>OP</font> used to create type of histogram (bar, barstacked, step, stepfilled). Default is bar.
- **align:** <font color='red'>OP</font> controls the plotting of histogram
- **weights:** <font color='red'>OP</font> contains array of weights having same dimensions as x
- **bottom:**	location of the baseline of each bin
- **rwidth:**	<font color='red'>OP</font> which is relative width of the bars with respect to bin width
- **color:**	<font color='red'>OP</font> used to set color of the plot. 
- **label:**	<font color='red'>OP</font> label of the plot
- **log:**	<font color='red'>OP</font> used to set histogram axis on log scale

```python
plt.figure(figsize=(9, 3))

#add transparency to the plots
plt.hist(bc_data[bc_data['diagnosis'] == 'M'].radius_mean, fc= (1, 0, 0), label = 'Malignant')
plt.hist(bc_data[bc_data['diagnosis'] == 'B'].radius_mean, fc= (0, 1, 0), label = 'Benign')
plt.legend()
plt.xlabel('Radius mean')
plt.ylabel('Frecuency')
plt.show()
```

## Boxplots

Syntaxis:

`plt.boxplot(data, notch=None, vert=None, patch_artist=None, widths=None)`

- **data:**	array or sequence of array to be plotted
- **notch:**	<font color='red'>OP</font> accepts boolean values
- **vert:**	<font color='red'>OP</font> accepts boolean values false and true for horizontal and vertical plot respectively
- **bootstrap:**	<font color='red'>OP</font> accepts int specifies intervals around notched boxplots
- **usermedians:**	<font color='red'>OP</font> accepts array or sequence of array dimension compatible with data
- **positions:**	<font color='red'>OP</font> accepts array and sets the position of boxes
- **widths:**	<font color='red'>OP</font> accepts array and sets the width of boxes
- **patch_artist:**	<font color='red'>OP</font> having boolean values
- **labels:**	sequence of strings sets label for each dataset
- **meanline:**	<font color='red'>OP</font> having boolean value try to render meanline as full width of box
- **order:**	<font color='red'>OP</font> parameter sets the order of the boxplot

```python
plt.figure(figsize=(9, 4))

d1 = bc_data[bc_data['diagnosis'] == 'M'].radius_mean
d2 = bc_data[bc_data['diagnosis'] == 'B'].radius_mean
data_box = [d1, d2]

plt.boxplot(data_box)
plt.xticks([1, 2], ['Malignant', 'Benign'])
plt.ylabel('Frecuency')
plt.xlabel('Radius mean')
plt.show()
```

```python
d1 = bc_data[bc_data['diagnosis'] == 'M'].radius_mean
d2 = bc_data[bc_data['diagnosis'] == 'B'].radius_mean
data_box = [d1, d2]

fig, ax = plt.subplots(figsize= (5, 5))
ax.boxplot(data_box)

ax.text(0.8, 16.8, "Mean",
        ha="center", va="center", rotation=20, size=8,
        bbox=dict(boxstyle="rarrow,pad=0.3",
        fc="lightblue", ec="steelblue", lw=2))

ax.text(1.8, 16.8, "Outliers",
        ha="center", va="center", rotation=0, size=8,
        bbox=dict(boxstyle="rarrow,pad=0.3",
        fc="lightblue", ec="steelblue", lw=2))

plt.xticks([1, 2], ['Malignant', 'Benign'])
plt.ylabel('Frecuency')
plt.xlabel('Radius mean')
plt.show()
```
