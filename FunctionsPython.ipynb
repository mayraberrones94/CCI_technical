### Classes and objects



Similar to the use we have of functions, we also have **Classes**. A class is basically a template of different functions, and (because of the concept of inheritance in Python) we can use a class without modifying it in different code

```
class MathFunctions:
    def add(self, x, y):
        """Returns the sum of x and y."""
        return x + y

    def subtract(self, x, y):
        """Returns the difference when y is subtracted from x."""
        return x - y

    def multiply(self, x, y):
        """Returns the product of x and y."""
        return x * y

    def divide(self, x, y):
        """Returns the quotient when x is divided by y."""
        if y == 0:
            return "Error: Division by zero is undefined."
        return x / y

    def power(self, base, exponent):
        """Returns the result of base raised to the power of exponent."""
        return base ** exponent

    def modulus(self, x, y):
        """Returns the remainder when x is divided by y."""
        if y == 0:
            return "Error: Division by zero is undefined."
        return x % y
```

```
MathFunctions.divide(None,3,10)
```

```
def read_image_custom(filename):
    """Reads an image file in binary mode and returns its raw data."""
    try:
        with open(filename, 'rb') as file:
            data = file.read()
        print("Image data read successfully.")
        return data
    except FileNotFoundError:
        print("Error: File not found.")
        return None

# Usage
image_data = read_image_custom("/content/drive/MyDrive/UAL_CCI/images/cat_meme.png")
print("Raw Image Data:", image_data[:100])  # Display the first 100 bytes as a preview

```

```
from PIL import Image
import matplotlib.pyplot as plt

def read_and_display_image(filename):
    """Reads and displays an image using Pillow and Matplotlib (for Jupyter notebook display)."""
    try:
        # Open the image file
        img = Image.open(filename)

        # Display image information
        print("Image Format:", img.format)
        print("Image Size:", img.size)
        print("Image Mode:", img.mode)

        # Display the image in the notebook
        plt.imshow(img)
        plt.axis('off')  # Turn off axis for a cleaner display
        plt.show()
    except FileNotFoundError:
        print("Error: File not found.")

# Usage
read_and_display_image("/content/drive/MyDrive/UAL_CCI/images/cat_meme.png")
```

## Data science
</div>

With Colab you can harness the full power of popular Python libraries to analyze and visualize data. The code cell below uses **numpy** to generate some random data, and uses **matplotlib** to visualize it. To edit the code, just click the cell and start editing.

You can import your own data into Colab notebooks from your Google Drive account, including from spreadsheets, as well as from Github and many other sources. To learn more about importing data, and how Colab can be used for data science, see the links below under [Working with Data](#working-with-data).

```
import numpy as np
import IPython.display as display
from matplotlib import pyplot as plt
import io
import base64

ys = 200 + np.random.randn(100)
x = [x for x in range(len(ys))]

fig = plt.figure(figsize=(4, 3), facecolor='w')
plt.plot(x, ys, '-')
plt.fill_between(x, ys, 195, where=(ys > 195), facecolor='g', alpha=0.6)
plt.title("Sample Visualization", fontsize=10)

data = io.BytesIO()
plt.savefig(data)
image = F"data:image/png;base64,{base64.b64encode(data.getvalue()).decode()}"
alt = "Sample Visualization"
display.display(display.Markdown(F"""![{alt}]({image})"""))
plt.close(fig)
```
